# Decimals - Precision & Representation

```Goal:```
Understand how decimals behave in computation and why precision matters.

Decimals are approximations of real numbers.

This **notebook** explores:

- Decimal precision
- Floating-point inaccuracies
- Visual comparison of close values

````python
print(0.1 + 0.2)

round(0.1 + 0.2, 2)
````

````python
import matplotlib.pyplot as plt

values = [0.1, 0.2, 0.3, 0.1 + 0.2]
labels = ["0.1", "0.2", "0.3", "0.1 + 0.2"]

plt.bar(labels, values)
plt.title("Decimal Precision Comparison")
plt.show()
````

**Observations:**

- Computers store decimals approximately.
- 0.1 + 0.2 isn't exactly 0.3.
- Precision must be handled carefully in programming and math.
