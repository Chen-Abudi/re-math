# Fractions - Visual & Numeric Exploration

```Goal:```
Understand how fractions map to decimals and why precision matters.

A fraction represents a division of two integers.

This **notebook** explores:

- How fractions convert to decimals
- Why some decimal repeat
- How fractions being compared visually

````python
from fractions import Fraction

fractions = [
  Fraction(1, 2),
  Fraction(1, 3),
  Fraction(1, 4),
  Fraction(2, 3),
  Fraction(3, 4),
]

for f in fractions:
    print(f"{f} = {float(f)}")
````

````python
import matplotlib.pyplot as plt  

values = [float(f) for f in fractions]
labels = [str(f) for f in fractions]

plt.bar(labels, values)
plt.xlabel("Fractions")
plt.ylabel("Decimal Value")
plt.title("Fractions as Decimal Values")
plt.show()
````

**Observations:**

- Fractions with denominators like 2 and 4 produce clean decimals.
- Fractions like 1/3 repeat infinitely.
- Visual comparison makes magnitude differences obvious.
